```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
```

```python
df = pd.read_csv('/content/MatchData.csv')
```

```python
df.head()
```

```python
df[['x', 'y']]=df.location.str.split(",", expand=True)
df[['endX', 'endY']]=df.pass_end_location.str.split(",", expand=True)
```

```python
df['x']= df['x'].str.replace("[", " ")
df['y']= df['y'].str.replace("]", " ")
df['endX']= df['endX'].str.replace("[", " ")
df['endY']= df['endY'].str.replace("]", " ")
```

```python
df['x']=df['x'].astype(float)
df['y']=df['y'].astype(float)
df['endX']=df['endX'].astype(float)
df['endY']=df['endY'].astype(float)
```

```python
df = df[df['type']=='Pass']
df = df[df['player']==name].reset_index()
```

```python
df['x']=df['x']*1.0
df['y']=df['y']*1.0
df['endX']=df['endX']*1.0
df['endY']=df['endY']*1.0
```

```python
df.head()
```

```python
from numpy import NaN
#fig, ax = plt.subplots(figsize=(13, 8.5))
#fig.set_facecolor('#22312b')
#ax.patch.set_facecolor('#22312b')

pitch = Pitch(pitch_type='statsbomb', pitch_color='#0a013b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#0a013b')
ax.patch.set_facecolor('#0a013b')

pitch.draw(ax=ax)
plt.gca().invert_yaxis()

for x in range(len(df['location'])):
  if df['pass_outcome'][x]!='Incomplete':
    plt.plot((df['x'][x], df['endX'][x]), (df['y'][x], df['endY'][x]), color='green')
    plt.scatter(df['x'][x], df['y'][x], color='green')
  if df['pass_outcome'][x]=='Incomplete':
    plt.plot((df['x'][x], df['endX'][x]), (df['y'][x], df['endY'][x]), color='red')
    plt.scatter(df['x'][x], df['y'][x], color='red')

plt.xlim(0, 120)
plt.ylim(0, 80)

plt.title(f'{name} Pass Map vs {awayTeam} on {date}', color='white', size=20, c='#39FF14')
```
